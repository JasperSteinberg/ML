Here we see very clearly why vessels might dissapear and reappear. AIS only has coverage of about 30km offshore, so if a vessel travels across large bodies of water, it will have large time gaps. We will use lagged versions of latitude and longitude as features, so it is important to understand their distribution. We will now clean up single-teleport event in the dataset, i.e. when a vessel travels more than 50 km/h in air-distance which is impossible, since shipping vessels have a max-speed of around 45 km/h. 

```python
#Calculate the distance between consecutive positions
def calculate_distance(row):
    if pd.isna(row['time_delta']):
        return np.nan  #First entry for each vessel
    prev_row = train.iloc[row.name - 1]
    if row['vesselId'] == prev_row['vesselId']:
        prev_coords = (prev_row['latitude'], prev_row['longitude'])
        curr_coords = (row['latitude'], row['longitude'])
        return geodesic(prev_coords, curr_coords).kilometers
    else:
        return np.nan

train['distance_km'] = train.apply(calculate_distance, axis=1)

#Calculate speed in km/h
train['speed_kmh'] = train['distance_km'] / train['time_delta']

#Define a realistic maximum speed for vessels (e.g., 50 km/h)
MAX_SPEED = 50

#Identify rows with unrealistic speeds
train['teleport'] = train['speed_kmh'] > MAX_SPEED
#After identifying teleporting entries

#Count the total number of teleporting instances
teleport_count = train['teleport'].sum()
print(f'Total number of teleporting instances: {teleport_count}')


#Remove teleporting entries
train = train[train['teleport'] != True].copy()

#Optionally, drop auxiliary columns
train.drop(columns=['time_delta', 'distance_km', 'speed_kmh', 'teleport'], inplace=True)
```

## Feature Engineering 
We create the decomposed velocity vectors for latitude and longitude. Since we train 1D-models, this aids in capturing the essence of movment within each coordinate. 

```python
train['v_lat'] = train['sog']*np.sin((np.pi/180)*train['cog'])
train['v_lon'] = train['sog']*np.cos((np.pi/180)*train['cog'])
```

Tree-based models dont learn temportal relationships in their natural state, so we will need to reformulate the problem. We want to predict vessel-positions up to 5 days into the future, whith only historical data available. Therefore, we divide the dataset into 5 day windows, and forward fill the last value for each vessel in each window into the next window. In essence we train our model to perform multi-step forcasting. We also create a time_diff variable, which is the time from the last seen observation to the current row. 

```python
#Calculate the minimum time for each vessel
train['vessel_min_time'] = train.groupby('vesselId')['time'].transform('min')

#Compute the time difference in seconds
train['time_diff_seconds'] = (train['time'] - train['vessel_min_time']).dt.total_seconds()

#Define the window duration in seconds
window_duration_seconds = 5 * 24 * 60 * 60  # 5 days in seconds

#Calculate 'window_id'
train['window_id'] = (train['time_diff_seconds'] // window_duration_seconds).astype(int)

#Drop the auxiliary columns
train.drop(columns=['vessel_min_time', 'time_diff_seconds'], inplace=True)
```

```python
#Sort the DataFrame
train_new = train.sort_values(by=['vesselId', 'window_id', 'time']).reset_index(drop=True).copy()

#Get the last observation per vessel and window
last_obs_per_window = train_new.groupby(['vesselId', 'window_id']).last().reset_index()

#Prepare and shift the last observations
#Select only the necessary columns to prevent overlap
last_obs_columns = [
    'vesselId', 'window_id', 'time', 'latitude', 'longitude',
    'v_lat', 'v_lon',
]
last_obs = last_obs_per_window[last_obs_columns].copy()

#Rename columns to distinguish them from 'train' columns
last_obs.rename(columns={
    'time': 'prev_window_time_last',
    'latitude': 'prev_window_last_latitude',
    'longitude': 'prev_window_last_longitude',
    'v_lat': 'prev_window_last_v_lat',
    'v_lon': 'prev_window_last_v_lon',
}, inplace=True)

shift_columns = [
        'prev_window_last_latitude', 'prev_window_last_longitude',
        'prev_window_time_last', 'prev_window_last_v_lat', 
        'prev_window_last_v_lon'
]

def shift_last_observations(df, shift_columns = shift_columns):
    df = df.sort_values('window_id').reset_index(drop=True)
    # Shift the columns by one to align with the next window
    
    df[shift_columns] = df[shift_columns].shift(1)
    return df

#Apply the shift per vesselId
last_obs_shifted = last_obs.groupby('vesselId').apply(shift_last_observations).reset_index(drop=True)

#Keep only the columns needed to merge
merge_columns = ['vesselId', 'window_id'] + shift_columns
last_obs_shifted = last_obs_shifted[merge_columns]


#Merge back into the original DataFrame
train_new = pd.merge(
    train_new,
    last_obs_shifted,
    on=['vesselId', 'window_id'],
    how='left'
)

#Compute 'time_diff'
train_new['time_diff'] = (train_new['time'] - train_new['prev_window_time_last']).dt.total_seconds()
```

We use the last window as a validation set, and will train on the rest. 

```python
#Find the maximum window_id
max_window_id = train_new['window_id'].max()
print(f"The maximum window_id (last window) is: {max_window_id}")
```

We validate using the next oldest window, as the oldest window dose not contain that many points. 

```python
#Create the validation set using the last window
validation_set = train_new[train_new['window_id'] == max_window_id-1].copy()
#validation_set_2 = train_new[train_new['window_id'] == max_window_id-1].copy()
#validation_set = pd.concat([validation_set_2, validation_set_1])

#Create the training set using all previous windows
training_set = train_new[train_new['window_id'] < max_window_id-1].copy()
```

Lastly, we set up our targets, features and datasets for model training

```python
#Define the target variables
targets = ['latitude', 'longitude']

#Select features
features = [
    'time_diff',                        # Time difference from previous window
    'prev_window_last_latitude',        # Last latitude from previous window
    'prev_window_last_longitude',       # Last longitude from previous window
    'prev_window_last_v_lat',   
    'prev_window_last_v_lon', 
]

#Combine features and targets for convenience
model_columns = features + targets

#Drop rows with missing values
training_set = training_set.dropna(subset=model_columns)
validation_set = validation_set.dropna(subset=model_columns)

```

## Testing different models

### LightGBM

Due to the large nature of the dataset, LightGBM is a cost-effective option. We prepare the training and validation sets. 

```python
X_train_lat = training_set[features]
y_train_lat = training_set['latitude']

X_valid_lat = validation_set[features]
y_valid_lat = validation_set['latitude']

X_train_lon = training_set[features]
y_train_lon = training_set['longitude']

X_valid_lon = validation_set[features]
y_valid_lon = validation_set['longitude']

#Latitude datasets
train_data_lat = lgb.Dataset(X_train_lat, label=y_train_lat)
valid_data_lat = lgb.Dataset(X_valid_lat, label=y_valid_lat)

#Longitude datasets
train_data_lon = lgb.Dataset(X_train_lon, label=y_train_lon)
valid_data_lon = lgb.Dataset(X_valid_lon, label=y_valid_lon)

```

We have pre-tranined the model

```python
#Define model parameters
params = {
    'objective': 'regression',   
    'metric': 'rmse',               
    'boosting_type': 'gbdt',
    'learning_rate': 0.05,
    'num_leaves': 31,
    'max_depth': -1,
    'min_data_in_leaf': 20,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'early_stopping_rounds': 50,
    'verbose': -1,
    'seed': 42
}
```

```python
#Train the latitude and longitude model
model_lat = lgb.train(
    params,
    train_data_lat,
    num_boost_round=1000,
    valid_sets=[valid_data_lat]
)

model_lon = lgb.train(
    params,
    train_data_lon,
    num_boost_round=1000,
    valid_sets=[valid_data_lon]
)
```

#### Model evaluation/interpretation

```python
#Predict on validation set
y_pred_lat = model_lat.predict(X_valid_lat, num_iteration=model_lat.best_iteration)

#Calculate evaluation metrics
rmse_lat = mean_squared_error(y_valid_lat, y_pred_lat, squared=False)
mae_lat = mean_absolute_error(y_valid_lat, y_pred_lat)
r2_lat = r2_score(y_valid_lat, y_pred_lat)

print(f"Latitude Model - Validation RMSE: {rmse_lat:.6f}")
print(f"Latitude Model - Validation MAE: {mae_lat:.6f}")
print(f"Latitude Model - Validation R^2 Score: {r2_lat:.6f}")
```

```python
#Predict on validation set
y_pred_lon = model_lon.predict(X_valid_lon, num_iteration=model_lon.best_iteration)

#Calculate evaluation metrics
rmse_lon = mean_squared_error(y_valid_lon, y_pred_lon, squared=False)
mae_lon = mean_absolute_error(y_valid_lon, y_pred_lon)
r2_lon = r2_score(y_valid_lon, y_pred_lon)

print(f"Longitude Model - Validation RMSE: {rmse_lon:.6f}")
print(f"Longitude Model - Validation MAE: {mae_lon:.6f}")
print(f"Longitude Model - Validation R^2 Score: {r2_lon:.6f}")
```

We can see that the R^2 score is quite good, even though the RMSE dosent look great. We now look at the gain and split importance plots 

```python
#Retrieve feature names from the model (assuming both models use the same features)
feature_names = model_lat.feature_name()

#Extract Gain Importance
gain_importance_lat = model_lat.feature_importance(importance_type='gain')
gain_importance_lon = model_lon.feature_importance(importance_type='gain')

#Extract Split Importance
split_importance_lat = model_lat.feature_importance(importance_type='split')
split_importance_lon = model_lon.feature_importance(importance_type='split')
```

```python
def plot_combined_importances(feature_names, gain_lat, gain_lon, split_lat, split_lon, top_n=20):
    """
    Plots Gain and Split Importances for Latitude and Longitude models side by side.
    
    Parameters:
    - feature_names (list): List of feature names.
    - gain_lat (list): Gain importance for latitude model.
    - gain_lon (list): Gain importance for longitude model.
    - split_lat (list): Split importance for latitude model.
    - split_lon (list): Split importance for longitude model.
    - top_n (int): Number of top features to display.
    """
    
    #Create DataFrames for better handling
    df_gain_lat = pd.DataFrame({'Feature': feature_names, 'Gain': gain_lat})
    df_gain_lon = pd.DataFrame({'Feature': feature_names, 'Gain': gain_lon})
    df_split_lat = pd.DataFrame({'Feature': feature_names, 'Split': split_lat})
    df_split_lon = pd.DataFrame({'Feature': feature_names, 'Split': split_lon})
    
    #Sort and select top_n features
    df_gain_lat = df_gain_lat.sort_values(by='Gain', ascending=False).head(top_n)
    df_gain_lon = df_gain_lon.sort_values(by='Gain', ascending=False).head(top_n)
    df_split_lat = df_split_lat.sort_values(by='Split', ascending=False).head(top_n)
    df_split_lon = df_split_lon.sort_values(by='Split', ascending=False).head(top_n)
    
    #Set up the matplotlib figure
    plt.figure(figsize=(20, 15))
    
    #Top Row - Gain Importance
    plt.subplot(2, 2, 1)
    sns.barplot(x='Gain', y='Feature', data=df_gain_lat, palette='Blues_d')
    plt.title('Latitude Feature Importance (Gain)', fontsize=16)
    plt.xlabel('Gain')
    plt.ylabel('Feature')
    
    plt.subplot(2, 2, 2)
    sns.barplot(x='Gain', y='Feature', data=df_gain_lon, palette='Greens_d')
    plt.title('Longitude Feature Importance (Gain)', fontsize=16)
    plt.xlabel('Gain')
    plt.ylabel('Feature')
    
    #Bottom Row - Split Importance
    plt.subplot(2, 2, 3)
    sns.barplot(x='Split', y='Feature', data=df_split_lat, palette='Blues_d')
    plt.title('Latitude Feature Importance (Split)', fontsize=16)
    plt.xlabel('Number of Splits')
    plt.ylabel('Feature')
    
    plt.subplot(2, 2, 4)
    sns.barplot(x='Split', y='Feature', data=df_split_lon, palette='Greens_d')
    plt.title('Longitude Feature Importance (Split)', fontsize=16)
    plt.xlabel('Number of Splits')
    plt.ylabel('Feature')
    
    plt.tight_layout()
    plt.show()
```

```python
#Define top_n features to display
top_n = 5  #Only train with 5 features

#Call the plotting function
plot_combined_importances(
    feature_names=feature_names,
    gain_lat=gain_importance_lat,
    gain_lon=gain_importance_lon,
    split_lat=split_importance_lat,
    split_lon=split_importance_lon,
    top_n=top_n
)
```

We see that the model places a large split importance on features that dont contribute mutch with respects to gain, which is not ideal. We can se though that the latitude model gets more out of the velocity vector for longitude and vice versa which is interesting. We tried using other features as well but the model picks up noise quite fast, and thus we choose to just stick with these 5. We could probably consider dropping v_lat for the latitude model and v_lon for the longitude model, which we will do when we retrain the model on the whole dataset. 

### XGBoost

```python
#Prepare the training and validation datasets for latitude
X_train_lat = training_set[features]
y_train_lat = training_set['latitude']

X_valid_lat = validation_set[features]
y_valid_lat = validation_set['latitude']

#Prepare the training and validation datasets for longitude
X_train_lon = training_set[features]
y_train_lon = training_set['longitude']

X_valid_lon = validation_set[features]
y_valid_lon = validation_set['longitude']

#Create DMatrix datasets for XGBoost - for latitude
train_data_lat = xgb.DMatrix(X_train_lat, label=y_train_lat)
valid_data_lat = xgb.DMatrix(X_valid_lat, label=y_valid_lat)

#Create DMatrix datasets for XGBoost - for longitude
train_data_lon = xgb.DMatrix(X_train_lon, label=y_train_lon)
valid_data_lon = xgb.DMatrix(X_valid_lon, label=y_valid_lon)
```

The model performes well without extensive tuning.

```python
def plot_combined_importances(feature_names, gain_lat, gain_lon, split_lat, split_lon, top_n=20):
    """
    Plots Gain and Split Importances for Latitude and Longitude models side by side.
    
    Parameters:
    - feature_names (list): List of feature names.
    - gain_lat (dict): Gain importance dictionary for latitude model.
    - gain_lon (dict): Gain importance dictionary for longitude model.
    - split_lat (dict): Split importance dictionary for latitude model.
    - split_lon (dict): Split importance dictionary for longitude model.
    - top_n (int): Number of top features to display.
    """
    
    #Convert importance dictionaries to DataFrames, filling missing features with zero importance
    df_gain_lat = pd.DataFrame(list(gain_lat.items()), columns=['Feature', 'Gain']).set_index('Feature').reindex(feature_names, fill_value=0).reset_index()
    df_gain_lon = pd.DataFrame(list(gain_lon.items()), columns=['Feature', 'Gain']).set_index('Feature').reindex(feature_names, fill_value=0).reset_index()
    df_split_lat = pd.DataFrame(list(split_lat.items()), columns=['Feature', 'Split']).set_index('Feature').reindex(feature_names, fill_value=0).reset_index()
    df_split_lon = pd.DataFrame(list(split_lon.items()), columns=['Feature', 'Split']).set_index('Feature').reindex(feature_names, fill_value=0).reset_index()
    
    #Sort and select top_n features
    df_gain_lat = df_gain_lat.sort_values(by='Gain', ascending=False).head(top_n)
    df_gain_lon = df_gain_lon.sort_values(by='Gain', ascending=False).head(top_n)
    df_split_lat = df_split_lat.sort_values(by='Split', ascending=False).head(top_n)
    df_split_lon = df_split_lon.sort_values(by='Split', ascending=False).head(top_n)
    
    #Set up the matplotlib figure
    plt.figure(figsize=(20, 15))
    
    #Top Row - Gain Importance
    plt.subplot(2, 2, 1)
    sns.barplot(x='Gain', y='Feature', data=df_gain_lat, palette='Blues_d')
    plt.title('Latitude Feature Importance (Gain)', fontsize=16)
    plt.xlabel('Gain')
    plt.ylabel('Feature')
    
    plt.subplot(2, 2, 2)
    sns.barplot(x='Gain', y='Feature', data=df_gain_lon, palette='Greens_d')
    plt.title('Longitude Feature Importance (Gain)', fontsize=16)
    plt.xlabel('Gain')
    plt.ylabel('Feature')
    
    #Bottom Row - Split Importance
    plt.subplot(2, 2, 3)
    sns.barplot(x='Split', y='Feature', data=df_split_lat, palette='Blues_d')
    plt.title('Latitude Feature Importance (Split)', fontsize=16)
    plt.xlabel('Number of Splits')
    plt.ylabel('Feature')
    
    plt.subplot(2, 2, 4)
    sns.barplot(x='Split', y='Feature', data=df_split_lon, palette='Greens_d')
    plt.title('Longitude Feature Importance (Split)', fontsize=16)
    plt.xlabel('Number of Splits')
    plt.ylabel('Feature')
    
    plt.tight_layout()
    plt.show()
```

```python
#Call the plotting function with the extracted importances
plot_combined_importances(
    feature_names=feature_names,
    gain_lat=gain_importance_lat,
    gain_lon=gain_importance_lon,
    split_lat=split_importance_lat,
    split_lon=split_importance_lon,
    top_n=top_n
)
```

### Predicting for the test data
We will use what we have learned to efficiently create the predictions in the short notebooks, one for lightGBM and one for XGBoost. We note that due to the randomness in the models, the kaggle score will probably change slightly.
